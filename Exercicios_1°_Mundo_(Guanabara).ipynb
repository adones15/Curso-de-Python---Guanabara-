17 - Faça um programa que leia o comprimento do cateto oposto e do cateto adjacente de um triângulo retângulo. Calcule e mostre o comprimento da hipotenusa.

```python
import math
cat_opos = float(input("Digite o valor do Cateto oposto: "))
cat_adja = float(input("Digite o valor do Cateto adjacente: "))

hipotenusa = math.hypot(cat_opos, cat_adja)

print("O valor do cateto oposto é {} e o cateto adjacente é {}, sendo assim o valor da hipotenusa é {:.2f}".format(cat_opos, cat_adja, hipotenusa))
```

18 - Faça um programa que leia um ângulo qualquer e mostre na tela o valor do seno, cosseno e tangente desse ângulo.

```python
import math

angulo = float(input())
seno = math.cos(math.radians(angulo))
cosseno = math.sin(math.radians(angulo))
tangente = math.tan(math.radians(angulo))
print("O ângulo de {}° tem o seno de {:.4f} o cosseno de {:.4f} e a tangente de {:.4f}".format(angulo, seno, cosseno, tangente))
```

```python
frase = "Curso em Video Python"

print(frase[9]) # Imprime apenas a letr/caractere correspondente a 9 no range
print(frase[7:18]) # Imprime todos os caracteres entre o range 7 e 17 (ignora o ultimo numero solicitado, no caso o 18)
print(frase[7:18:2]) # imprime todos os caracteres do range igual no ultimo exemplo, porém pulando de 2 em 2
print(frase[7:18:3]) #imprime todos os caracteres do range igual no ultimo exemplo, porém pulando de 3 em 3
print(frase[:5]) # quando não coloca nenhum numero e coloca somente o ":" e o numero, indica que ele vai pegar os caracteres do 0 até o 5
print(frase[5:]) # mesma coisa do ultimo exemplo, ele vai pegar do caractere 5 em diante até o ultimo, quando se coloca o ":" e não especifica, ele pega 
#todos os caracteres, ou antes ou depois do numero
print(frase[9::3]) # aqui ele começa no 9, como entre os ":" está vazio, ele vai até o ultimo caracter do rang, porém pulando de 3 em 3
```
